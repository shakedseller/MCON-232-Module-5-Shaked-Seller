## Building our CSV file named "student_data.csv"

````python
data = ["name", "age", "major", "course", "grade",
        "Alice", 20, "CS", "Math", 90,
        "Alice", 20, "CS", "CS", 88,
        "Bob", 21, "Math", "Math", 85,
        "Bob", 21, "Math", "CS", 90]

# WRITE
count = 1
with open("student_data.csv", "w") as fout:
    for item in data:
        item = str(item)
        if count == len(data) :
            fout.write(item)
        elif count % 5 == 0:
            fout.write(item + "\n")
        else:
            fout.write(item + ",")   # cleaner CSV
        count += 1

# Read this in
# 1st row - column names
# between each data there is comma except for end (newline)
# use column names as keys

student_list = {}

with open("student_data.csv", "r") as filename:
  header_parts = [f.strip() for f in filename.readline().strip().split(",")]

  for line in filename:
    parts = [p.strip() for p in line.strip().split(",")]
    name, age, major, course, grade = parts
    for p in parts:
      p.strip("\n")

    if name not in student_list:
      student_list[name] = {header_parts[1] : parts[1],
                            header_parts[2] : parts[2],
                            "grades" : {},}

    student_list[name]["grades"][course] = grade

  print(student_list)
````

````python
def add_student(school, student, age, major):
  if not isinstance(school, dict):
    raise TypeError("school must be a dictionary")

  if not isinstance(student, str) or student.strip() == "":
    raise ValueError("student must be a non-empty string")

  if not isinstance(age, int) or not 0 < age <= 25:
    raise ValueError("age must be an int in the range (0 - 25)")

  if not isinstance(major, str) or major.strip() == "":
    raise ValueError("major must be a non-empty string")

  if student not in school:
    school[student] = {"age": age, "major": major, "grades" : {}}
    return

  if not isinstance(school[student], dict):
    raise TypeError(f"{student} value must be a dictionary")

  if "age" not in school[student]:
    school[student]["age"] = age

  if "major" not in school[student]:
    school[student]["major"] = major

  if "grades" not in school[student]:
    school[student]["grades"] = {}

  if not isinstance(school[student]["grades"], dict):
    raise TypeError(f"grades for {student} must be a dictionary")

  print(f"{school[student]} has been added")

add_student(school, "Someone", 19, "CS")
````

````python
# Return all letters that are unique
def exactly_once(items):
  answer = []
  unseen = set(items)

  for i in items:
    if i in unseen:
      unseen.discard(i)
      answer.append(i)
    else:
      if i in answer:
        answer.remove(i)

  return answer

print(exactly_once(["a", "b", "c", "c", "c", "c", "c", "d", "d"]))
````

## Leetcode 217 :
Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

````python
# answer
def twice_value(nums):
  return len(set(nums)) != len(nums)

print(twice_value([1, 2, 2, 3]))
print(twice_value([1, 2, 3]))
````

## Leetcode 202 - Happy Number
Write an algorithm to determine if a number n is happy.

A happy number is a number defined by the following process:

Starting with any positive integer, replace the number by the sum of the squares of its digits.
Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
Those numbers for which this process ends in 1 are happy.
Return true if n is a happy number, and false if not.



Example 1:

- Input: n = 19
- Output: true
- Explanation:
- 1^2 + 9^2 = 82
- 8^2 + 2^2 = 68
- 6^2 + 8^2 = 100
- 1^2 + 0^2 + 02 = 1
Example 2:

Input: n = 2
Output: false


Constraints:

1 <= n <= 231 - 1

````python
# solution
def happy_number(n):
  seen = set()

  #seen.add(sum(int(d) ** 2 for d in str(n)))
  while n != 1 and n not in seen:
    seen.add(n)
    n = sum(int(d) ** 2 for d in str(n))

  return n == 1

print(happy_number(19))
print(happy_number(2))
````
